```python
def plot_corr(df, size=20):
    """
    Function plots a graphical correlation matrix for each pair of columns in the dataframe.

    Input:
        df: pandas DataFrame
        size: vertical and horizontal size of the plot

    Displays:
        matrix of correlation between columns.  Blue-cyan-yellow-red-darkred => less to more correlated
                                                0 ------------------>  1
                                                Expect a darkred line running from top left to bottom right
    """

    corr = df.corr()    # data frame correlation function
    fig, ax = plt.subplots(figsize=(size, size))
    ax.matshow(corr)   # color code the rectangles by correlation value
    plt.xticks(range(len(corr.columns)), corr.columns)  # draw x tick marks
    plt.yticks(range(len(corr.columns)), corr.columns)  # draw y tick marks
```

# IMPROVING ACCURACIES


## 1. FINDING RANDOM_STATE VALUE FOR LOGISTIC REGRESSION

```python
r_start=1
r_end=5000
r_inc=1

r_values,accracy_scores =[],[]

r_val=r_start

best_recall_score = 0
best_accuracy_score=0
best_r_value=0

  
while (r_val<r_end):
    r_values.append(r_val)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=r_val)
    lr_model_loop = LogisticRegression(random_state=r_val)#,class_weight="balanced")

    lr_model_loop.fit(X_train, y_train.ravel())
    lr_predict_loop_test = lr_model_loop.predict(X_test)

    accuracy_score= metrics.accuracy_score(y_test,lr_predict_loop_test)
    accracy_scores.append(accuracy_score)

    if (accuracy_score >best_accuracy_score):
        best_accuracy_score = accuracy_score
        best_lr_predict_test = lr_predict_loop_test
        best_r_value = r_val
        
    r_val = r_val + r_inc


print("best_random_state_value: ",best_r_value)
print("best_accuracy_score: ",best_accuracy_score)

%matplotlib inline 

fig=plt.figure(figsize=(15,5))
plt.plot(r_values,accracy_scores, "-")
plt.xlabel("random_state_value")
plt.ylabel("accuracy_score")

str="acc_scoreVSrs_value__LRegression1"

plt.savefig('/home/adeeshwar/PythonWorlspace/'+str+'.png')
```
